```python
#1
from transformers import pipeline

# Initialize the default question-answering pipeline
qa_pipeline = pipeline("question-answering")

# Provide context and question
context = "Charles Babbage is considered the father of the computer. He designed the Analytical Engine in the 19th century."
question = "Who is considered the father of the computer?"

# Get the answer
result = qa_pipeline(question=question, context=context)

# Print the result
print("Answer:", result['answer'])
print("Score:", result['score'])
print("Start Index:", result['start'])
print("End Index:", result['end'])

```

```python
#2
from transformers import pipeline

# Initialize the QA pipeline with a custom pretrained model
qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")

# Provide context and question
context = "Charles Babbage is considered the father of the computer. He designed the Analytical Engine in the 19th century."
question = "Who is considered the father of the computer?"

# Get the answer
result = qa_pipeline(question=question, context=context)

# Print the result
print("Answer:", result['answer'])
print("Score:", result['score'])
print("Start Index:", result['start'])
print("End Index:", result['end'])
```

```python
from transformers import pipeline

# Load custom QA model
qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")

# Custom context
context = """
Senthil is pursuing a master's degree in Data Science and Artificial Intelligence at the University of Central Missouri.
He has more than 20 years of experience in teaching Engineering subjects.
"""

# Question 1
question1 = "What is Senthil studying?"
result1 = qa_pipeline(question=question1, context=context)

# Question 2
question2 = "How many years of experience does Senthil have?"
result2 = qa_pipeline(question=question2, context=context)

# Print results
print("Question 1:", question1)
print("Answer 1:", result1['answer'])
print("Score 1:", result1['score'])

print("\nQuestion 2:", question2)
print("Answer 2:", result2['answer'])
print("Score 2:", result2['score'])
```

```python
#Short Answer 1: How does a Conditional GAN differ from a vanilla GAN?
A Conditional GAN (cGAN) is different from a vanilla GAN in one key way:

✅ In a cGAN, both the Generator and Discriminator receive extra information — such as class labels — along with their normal inputs.

This means:

The Generator learns to create images conditioned on a specific label (e.g., generate a "7").

The Discriminator not only decides if an image is real or fake, but also checks whether the image matches the given label.

📌 In contrast, a vanilla GAN has:

No label input.

The Generator produces images from random noise only.

The Discriminator just classifies real vs fake, not based on any specific class.

✅ Real-World Application where conditioning is important:
🎯 Face generation with attributes:
Generate faces with specific features (e.g., "a smiling woman with glasses"). Here, conditioning on attributes like age, gender, expression, or accessories allows fine control.

Example: In AI art tools or fashion try-on systems, users can generate images with selected features.

✅ Short Answer 2: What does the discriminator learn in an image-to-image GAN?
In an image-to-image GAN (like Pix2Pix), the Discriminator learns to:

✅ Distinguish whether an output image is a realistic translation of the input image.

So, it evaluates paired inputs, like:

A sketch → photo

A grayscale image → color version

A segmentation map → street view photo

The Discriminator checks:

Is this a realistic image?

Does it match the input it's paired with?

✅ Why is pairing important in this context?
🔗 Pairing is essential because the GAN is learning a mapping from one image domain to another.

Without correct pairs:

The model can't learn the proper input-output relationship.

The output might be realistic, but unrelated to the input.

Example: If a shoe sketch is paired with a wrong shoe photo, the Generator learns a wrong mapping, which ruins training.

```
